```python
import numpy as np
x_train=np.array([1.0, 2.0])
y_train=np.array([100, 300])
print(f"x_tain={x_train}")
print(f"y_tain={y_train}")
```

showning the data in graph form

```python
import matplotlib.pyplot as plt
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("housing prices")
plt.ylabel("price")
plt.xlabel("size")
plt.show()
```

assuming the weight and biasness of the function


```python
w=200
b=-100
```

```python
def compute_model_output(x, w, b):
    m=x.shape[0]
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]=w*x[i] + b
    return f_wb
```

```python
temp_wb=compute_model_output(x_train, w, b)
plt.scatter(x_train, y_train, label="actual value", marker="x", c="r")
plt.plot(x_train, temp_wb, label="prediction" ,c='b')
plt.title("house prices prediction")
plt.ylabel("house price")
plt.xlabel("house size")
plt.legend()
```
